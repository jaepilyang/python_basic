```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time
```

```python
search_text = input('크롤링할 키워드를 입력하세요.')
```

```python
driver = webdriver.Chrome('./chromedriver') # working directory에 있으면 따로 경로지정 안 해두 됨
driver.set_window_size(800, 1000)
driver.get('https://korean.visitkorea.or.kr')
time.sleep(2)
```

```python
driver.find_element_by_id('chkForm01').click() # 팝업 닫기
elem = driver.find_element_by_id('inp_search')# 검색창 접근
elem.send_keys(search_text) # 검색창 입력
driver.find_element_by_class_name('btn_search2').click() # 검색버튼 입력
time.sleep(2)
```

```python
page_num = 6
f = open('test.txt', 'a', encoding = 'utf-8')

for i in range(1, page_num+1):
    html_source = driver.page_source
    soup = bs(html_source, 'html.parser')
    
    content_list = soup.select('#contents > div > div.box_leftType1 > ul')
    lis = content_list[0].select('li')
    
    for li in lis:
        print(li.text)
        print()
        f.write(li.text + '\n')
        
    i += 1
    print(i)
    time.sleep(2)
    try:
        driver.find_element_by_link_text('{}'.format(i)).click()
    except:
        driver.find_element_by_link_text('다음').click()
    time.sleep(2)
    
    if i > page_num+1:
        f.close()
        driver.quit()
        break
```

```python
def visitkor_search():
    search_text = input('크롤링할 키워드 입력: ')
    driver = webdriver.Chrome('./chromedriver')
    driver.set_window_size(800, 1000)
    driver.get('https://korean.visitkorea.or.kr')
    time.sleep(3)
    driver.find_element_by_id('chkForm01').click()

    elem = driver.find_element_by_class_name('inp_search')
    elem.send_keys(search_text)
    driver.find_element_by_class_name('btn_search2').click()
    html = driver.page_source
    soup = bs(html, 'html.parser')
    content_list = soup.select('#contents > div > div.box_leftType1 > ul')
    lis = content_list[0].select('li')

    for i in range(0,len(lis),3):
        print('제목: ', lis[i].select('div.tit > a')[0].text)
        print('지역: ', lis[i].select('div.area_txt > p')[0].text)
        print('태그: ', lis[i].select('div.area_txt > p > span')[0].text)
        print('='*40)
    
    title_list = []
    area_list = []
    tag_list = []
    
    for i in range(0,len(lis),3):
        title_list.append(lis[i].select('div.tit > a')[0].text)
        area_list.append(lis[i].select('div.area_txt > p')[0].text)
        tag_list.append(lis[i].select('div.area_txt > p > span')[0].text)
        
    
    time.sleep(3)
    driver.quit()
    
    return title_list, area_list, tag_list, driver
```

```python
page_num = int(input('몇 페이지까지 크롤링할지 입력'))

title_list = []
area_list = []
tag_list = []

for i in range(2, page_num+1):
    title_list, area_list, tag_list, driver = visitkor_search()
    
    title_list.append(title_list)
    area_list.append(area_list)
    tag_list.append(tag_list)
    
    driver.find_element_by_id('{}'.format(i)).click()
```
