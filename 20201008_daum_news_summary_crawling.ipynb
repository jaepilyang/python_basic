```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://search.daum.net/search?w=news&nil_search=btn&DA=NTB&enc=utf8&cluster=y&cluster_page=1&q=%EB%B9%85%EB%8D%B0%EC%9D%B4%ED%84%B0'
html = requests.get(url)
print(html.text)
```

```python
# li.fst > div.wrap_cont > div > div.wrap_tit.mg_tit > a
# li.fst > div.wrap_cont > div > div.wrap_tit.mg_tit >a
# li.fst > div.wrap_cont > div > span.f_nb.date
# li.fst > div.wrap_cont > div > span.f_nb.date
# li.fst > div.wrap_cont > div > p ')
soup = bs(html.text, 'html.parser')
```

```python
print(soup.select('li.fst > div.wrap_cont > div > div.wrap_tit.mg_tit > a')[0].text)
print()
print(soup.select('a.f_link_b')[0].text)
# a에 f_link_b 클래스로 있었네...
```

```python
res = soup.select('a.f_link_b') # res로 url과 제목 뽑아냄

for t in res:
    print(t.text)
    print()
    
for u in res:
    print(u['href'])
```

```python
pt = soup.select('span.f_nb.date') #여기에서 언론사와 날짜 뽑아낼 것 줄여서 pt

for p in pt: # 언론사
    print(p.text.split()[2])

print()
    
for d in pt: # 날짜
    print(d.text.split()[0])
```

```python
su = soup.select('div.wrap_cont > div > p') # 여기는 요약 부분

for s in su:
    print(s.text)
    print()
```
