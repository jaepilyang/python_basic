```python
# 주소록 제작 프로젝트

# 1. 주소록 입력 - 이름, 전화번호, 이메일 / 입력 포맷
# 2. 주소록 출력
# 3. 주소록 저장
# 4. 주소록 삭제
# 5. 주소록 파일을 제거
```

```python
import os.path
import pickle
```

```python
# 연락처 클래스 - 인스턴스를 생성해서 사용

class Contact:
    # 생성자 함수
    def __init__(self, name, phone_num, email):
        self.name = name #인스턴스 변수 : self.~
        self.phone_num = phone_num
        self.email = email
        
    # 연락처를 출력 메소드 - 인스턴스메소드(self를 인자로 받는다.)
    def prt_info(self):
        print('Name: {}'.format(self.name))
        print('Phone_Number: {}'.format(self.phone_num))
        print('E-mail: {}'.format(self.email))
        print('='*30)
        print()
```

```python
# 연락처 정보 입력 클래스 인스턴스 형식으로 저장

def add_cont(c_list):
    name = input_name()
    phone = input_phone()
    email = input_email()
    print()
    
    # 이름, 전화, 주소를 클래스 객체를 만들어서 객체를 리스트에 담아서 저장
    for v in c_list: # c_list는 주소록을 담은 인스턴스들을 요소로 가진 리스트
        if name == v.name: # 이전의 인스턴스 이름과 동일한 이름이 있는지 확인
            print('Name exists!!')
            print()
            break
    else:
        cont = Contact(name, phone, email) # 이름을 입력받고 클래스 객체 생성
        print('Saved!')
        print()
        c_list.append(cont)
```

```python
# 이름으로 조회된 연락처를 삭제 == 이름으로 조회된 인스턴스를 삭제
# c_list : 주소록을 입력받은 후 인스턴스로 만들어서 그 인스턴스들을 담은 리스트
#             주소들이 객체화되어 c_list에 담겨있음

def del_cont(c_list): # 인스턴스가 들어가 있는 리스틑를 매개변수로 받음
    nm = input('Name : ') # 삭제할 이름을 입력받고
    print()
    
    if len(c_list) > 0:
        for i, cont in enumerate(c_list): # 주소록 인스턴스를 하나씩 가져옴
            if str(cont.name).strip() == nm.strip(): # strip() - 문자열 좌우 공백 제거
                print("'{}' deleted.".format(cont.name))
                print()
                del c_list[i]
                break
        else:
            print('No files to delete')
    else:
        print('No files to delete')
```

```python
# 저장된 모든 연락처 정보를 출력
# c_list에 연락처 객체를 담고있는 모든 객체를 가져와서 prt_info()메소드를 
# 통해서 연락처 정보를 출력

def prt_cont(c_list):
    if len(c_list)>0:
        for i in c_list:
            i.prt_info()
    else:
        print('Contact is empty!!')
```

```python
# 주소록이 객체가 담긴 리스트 c_list를 파일로 저장
# => 파이썬 객체를 파일로 저장하는 pickle함수 이용
# with open('파일명', 'wb') as f:
#             pickle.dump(저장할 객체, f)
        
def store_cont_db(c_list):
    try:
        with open('cont_db.bin', 'wb') as f:
            pickle.dump(c_list, f) # pickle: 파이썬 객체를 파일로 저장하는 함수
            print('db daved')
            print()
    except IOError as log: # IOError: 파일을 열지 못할 때 발생하는 오류
        print(log)
```

```python
# 파일 db 업로드

def load_cont_db():
    p_list = []
    if os.path.isfile('cont_db.bin'): # 현재 dir에 'cont_db.bin' 파일이 있으면 True
        try:
            with open('cont_db.bin', 'rb') as f:
                p_list = pickle.load(f) # pickle.load() 함수로 pickle로 저장된 파일을
        except IOError as log: # 파일을 열지 못하는 error에 대해 예외처리
            print(log)
    else: # 파일이 없으면
        print('DB File not found!')
        print()
    return p_list # pickle로 열어서 가져온 리스트를 리턴
```

```python
# 파일 DB 삭제

def drop_cont_db():
    if os.path.islife('cont_db.bin'):
        try:
            os.remove('cont_db.bin')
            print('DB File dropped')
        except FileNotError as log:
            print(log)
    else:
        print('DB File not found!!')
        print()
```

```python
# 이름 입력

def input_name():
    while True:
        try:
            name = input('Name: ')
            if len(name)<2: # 두 글자 이상의 이름을 받기 위해
                raise ValueError # 강제로 오류 발생
            else:
                break
        except ValueError:
            print('Name is too short!!')
    return name
```

```python
# 전화번호 입력
def input_phone():
    while True:
        try:
            phone_num = input('Phone number: ')
            if len(phone_num) < 7:
                raise ValueError
            else:
                break
        except ValueError:
            print('Phone Number is too short')
    return phone_num
```

```python
# 이메일 입력
def input_email():
    while True:
        try:
            email = input('E-mail: ')
            if len(email) < 5:
                raise ValueError
            else:
                break
        except ValueError:
            print('E-mail is too short')
    return email
```

```python
# 이용자에게 메뉴 출력
def prt_menu():
    print('1. Add') # 주소록 입력
    print('2. Info') # 주소록 출력
    print('3. Delete') # 입력데이터 삭제
    print('4. DB Saved') # 파일 DB로 저장
    print('5. DB Drop') # 파일 DB 삭제
    print('6. Exit') # 나가기
    print()
    
    menu = input('Select Menu Number')
    print()
    return int(menu)
```

```python
# 프로그램 실행

def main():
    c_list = load_cont_db() # 주소 객체를 담은 file db를 가져와서 리스트로 받음
    
    while True:
        menu = prt_menu() # 이용ㅇ자에게 입력 메뉴를 받음
        if menu == 1:
            add_cont(c_list) # 주소록 입력 : 주소록 객체 생성 - c_list에 저장
        elif menu == 2:
            prt_cont(c_list) # 주소록 출력 : c_list에서 일치하는 이름 객체 출력
        elif menu == 3:
            del_cont(c_list) # 입력데이터 삭제 : c_list에서 일치하는 데이터 삭제
        elif menu == 4:
            store_cont_db(c_list) # 주소록 DB 파일로로 저장
        elif menu == 5:
            drop_cont_db(c_list) # DB 파일 삭제
        else:
            break # 다른 입력인 경우 빠져나감
```

```python
main()
```
