```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
res = requests.get('https://movie.naver.com/movie/point/af/list.nhn')
res.status_code

html = res.text
soup = bs(html, 'html.parser')
soup
```

```python
for p in range(10):
    res = requests.get('https://movie.naver.com/movie/point/af/list.nhn?&page={}'.format(p+1))
    html = res.text
    soup = bs(html, 'html.parser')
    soup

    print('{}페이지'.format(p+1))
    for i in range(10):
        subject = soup.select('a.movie.color_b')[i].text
        star = soup.select('div.list_netizen_score > em')[i].text
        review = soup.select('td.title')[i].text.split('\n')[5]
    
        print('{}번, '.format(i+1), end=' ')
        print('{}, '.format(subject), end=' ')
        print('{}, '.format(star), end=' ')
        print('{}'.format(review))
    print()
        
        
#         title_list.append(title)
#         score_list.append(title)
#         genre_list.append(title)
#         actors_list.append(title)
```
