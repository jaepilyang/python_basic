```python
# 태크에 효과 입히기 효과 부여
# test3.html

<html>
    <head>
        <style>
            p {color:blue;}
            div {color:red;}
        </style>

    </head>

    <body>
        <p> paragraph1 </p>
        <p> paragraph1 </p>
        
        <div> paragraph3 </div>

    </body>
</html>
```

```python
# css를 통해서 태그에 효과 입히기 - id나 클래스 의미
# test4.html

<html>
    <head>
        <style>
            #title{color:blue;} #id를 통한 태그에 효과주기 - 한 곳에만 적용가능
            .par {color:red;} # class를 통한 태그에 효과주기 - 여러곳에 적용가능
        </style>

    </head>

    <body>
        <p id = 'title'> paragraph1 </p> # title id를 적용
        <p class = 'par'> paragraph1 </p> # par class를 적용
        
        <div class = 'par'> paragraph3 </div> # par class를 적용

    </body>
</html>
```

```python
from bs4 import BeautifulSoup as bs
import requests # 웹사이트의 html 스크립트를 긁어오는 라이브러리

url = 'https://www.naver.com/'

html = requests.get(url)
print(html)
print(html.text)
```

```python
#NM_FAVORITE > div.group_nav > ul.list_nav.NM_FAVORITE_LIST > li
```

```python
soup = bs(html.text, 'html.parser')
```

```python
res = soup.select('#NM_FAVORITE > div.group_nav > ul.list_nav.NM_FAVORITE_LIST > li>a')
res[0]
```

```python
for t in res:
    print(t.text)
```

```python
for t in res:
    print(t['href'])
```

```python
for t in res:
    print(t.text, ':', t['href'])
```
