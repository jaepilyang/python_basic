```python
# 년도별 남아 신생아 수와 여아 신생아 수를 각각 출력

for i in range(1880,2016):
    with open('./names/yob{}.txt'.format(i), 'r') as f:
        text = f.readlines()
        sumf = 0
        summ = 0
        for l in text:
            if l.split(',')[1] == 'F':
                sumf += int(l.split(',')[2])
            else:
                summ += int(l.split(',')[2])
        print('{}년도 남아출생아수: {}, 여자출생아수{}'.format(i, summ, sumf))
```

```python
# 연도별로 인기있는 상위 10개 이름을 성별로 출력
import operator

for i in range(1880,2015):
    with open('./names/yob{}.txt'.format(i), 'r') as f:
        text = f.readlines()
        dic_f = {}
        dic_m = {}
        top_f = []
        top_m = []
        for l in text:
            num = l.split(',')[2]
            gen = l.split(',')[1]
            nam = l.split(',')[0]
                        
            if gen == 'F':
                dic_f[nam] = int(num)
            else:
                dic_m[nam] = int(num)
            
        dic_fs = sorted(dic_f.items(), key=lambda x: x[1], reverse=True)
        dic_ms = sorted(dic_m.items(), key=lambda x: x[1], reverse=True)
        
        for c in dic_fs[:10]:
            top_f.append(c[0])
        for c in dic_ms[:10]:
            top_m.append(c[0])
    print('연도: {}'.format(i))
    print('남자', top_m)
    print('여자', top_f)
    print('='*25)
```

```python
path = "D:\\BigData\\python\\names\\"
file_path = path + "yob{}.txt"

for y in range(1880, 2016):
    file_path1 = file_path.format(y)
    
    with open(file_path1, 'r') as f:
        data = f.readlines()
        fbn = {}
        mbn = {}
        for l in data:
            bp = l.split(',')[2]
            bg = l.split(',')[1]
            bn = l.split(',')[0]

            if bg == 'F':
                fbn[bn] = int(bp)
            else:
                mbn[bn]= int(bp)
                
        fbs = sorted(fbn.items(), key=lambda x: x[1], reverse=True)
        mbs = sorted(mbn.items(), key=lambda x: x[1], reverse=True)
        
        fns = []
        for i in fbs[:10]:
            fns.append(i[0])
            
        mns = []
        for i in mbs[:10]:
            mns.append(i[0])
            
    print('연도:', y)
    print('남아 10:', mns)
    print('여아 10:', fns)
    print('='*30)
```

```python
# 고유방문자 수와 총 방문횟수
with open('./files/access_log', 'r') as f:
    logs = f.readlines()
    c = {}
    count = 0
    acc = 0
    for l in logs:
        id = l.split()[0]
        if id in c.keys():
            c[id] += 1
        else:
            c[id] = 1
            count += 1
    
    L = []
    L = c.values()
    acc = sum(L)
    
    print('총 방문자수: {}'.format(count))
    print('총 방문횟수: {}'.format(acc))
    
```
