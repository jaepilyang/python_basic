```python
# 메모장 만들기
# 메모할 내용을 입력받은 후, 파일에 'memo.txt'이 누적해서 기록하고 동시에
# 전체 메모내용을 출력하는 코드 작성

# 모드 'w' => 덮어쓰기 모드
# 모드 'a' => 이어쓰기
```

```python
text = input('메모할 내용 입력: ')

f = open('memo.txt', 'a')
f.write(text)
f.close()

f = open('memo.txt', 'r')
lines = f.read()
f.close()

print()
print('메모장 내용\n', '='*20)
print(lines)
```

```python
# 1. 'stockcode.txt' 를 dict 형태로 저장 {'삼성전자':0234548, }
# 2. 주가이름을 입력하면 바로 종목코드를 얻을 수 있도록
```

```python
f = open('stockcode.txt', 'r')
stock = {}
for line in f:
    a = line.split()[1]
    b = line.split()[0]
    stock[a] = b
print(stock)
f.close()
```

```python
stock['삼일제약']
```

```python
f = open('stock_history.txt', 'a')
text = input('종목명을 입력하시오: ')
f.write(text+' ')
print(stock[text])
f.close()
```
